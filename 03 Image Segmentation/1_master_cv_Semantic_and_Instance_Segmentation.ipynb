## Code Example 5.1 (i) Synthetic Image Generation for Semantic Segmentation Tasks

Short Description: This code creates a controlled, synthetic image dataset specifically designed for teaching semantic and instance segmentation concepts. It generates a simplified urban scene containing multiple objects of different classes, enabling focused study of segmentation algorithms without the complexity of real-world imagery.

Key Libraries Used:

    PIL.Image and PIL.ImageDraw: Core tools for programmatically creating and drawing on images. Image.new() creates blank canvases, while ImageDraw.Draw() provides methods to draw shapes.

    numpy: Converts the final PIL Image object into a NumPy array, which is the standard format for most computer vision and machine learning libraries.

    matplotlib.pyplot: Displays the generated synthetic image for visual verification.

Code Logic and Flow: High-Level Overview

This script addresses a common challenge in computer vision education: obtaining suitable, controlled data for segmentation tasks. Instead of using complex real-world images, it programmatically generates a simple synthetic scene. The logic follows a painterly approach, starting with a blank canvas and sequentially adding background elements (sky, road) and foreground objects (multiple cars, trees). Each object is drawn as a combination of simple geometric shapes (rectangles, ellipses) with distinct colors and positions, simulating different semantic classes and instances. This results in a clean, well-defined image ideal for demonstrating how segmentation algorithms distinguish between different object types and individual object instances.

Visual Flowchart



Step-by-Step Code Breakdown
Step 1: Canvas Initialization and Background Drawing

    Base Image Creation: The function create_sample_image starts by creating a new PIL Image object with specified width and height (default 400x300). The color='lightblue' parameter fills the entire canvas, simulating a sky background.

    Drawing Context: An ImageDraw.Draw object is created from the image, providing methods to add shapes and lines.

    Background Layer: A gray rectangle is drawn from the horizontal center to the bottom of the image ([0, 200, width, height]), representing a road. This establishes the primary background class against which objects will be segmented.

Step 2: Drawing Object Class 1: Multiple Car Instances

    Instance-Based Drawing: Three distinct car instances are drawn, each with a unique position and color (red, blue, green). This is crucial for teaching instance segmentation, where the goal is to separate individual objects within the same class.

    Hierarchical Structure: Each car is composed of multiple parts:

        A main colored rectangle for the car body (fill='red', 'blue', 'green').

        Two smaller 'lightblue' rectangles on each car body representing windows. This creates a simple internal structure.

    Spatial Variation: The cars are placed at different vertical coordinates (180, 170, 185), making them appear at different positions relative to the road, adding spatial complexity.

Step 3: Drawing Object Class 2: Tree Instances

    Multi-Shape Objects: Two tree instances are drawn, each combining two different shapes and colors to form a single object:

        A 'darkgreen' ellipse represents the tree's foliage.

        A 'brown' rectangle below it represents the trunk.

    Class Differentiation: The trees use colors and shapes distinct from the cars, clearly defining a second semantic class for semantic segmentation tasks, where the goal is to classify each pixel by its object category.

Step 4: Finalization and Output

    Format Conversion: The function concludes by converting the PIL Image object to a NumPy array using np.array(img). This is a critical step because most computer vision libraries (OpenCV, PyTorch, TensorFlow) process images as NumPy arrays.

    Visual Verification: The generated image is displayed using matplotlib with a descriptive title. Turning the axis off (plt.axis('off')) provides a clean view.

    Property Inspection: The code prints the image's shape (e.g., (300, 400, 3) for height, width, RGB channels) and dtype (typically uint8), confirming the array's structure is correct for downstream processing.

```python
# 3. Create Synthetic Image Data for Segmentation

def create_sample_image(width=400, height=300):
    """Create a synthetic image with multiple objects"""
    img = Image.new('RGB', (width, height), color='lightblue')
    draw = ImageDraw.Draw(img)

    # Draw road (background)
    draw.rectangle([0, 200, width, height], fill='gray')

    # Draw multiple cars (instances of same class)
    # Car 1
    draw.rectangle([50, 180, 120, 220], fill='red', outline='darkred', width=2)
    draw.rectangle([60, 190, 80, 210], fill='lightblue')  # window
    draw.rectangle([90, 190, 110, 210], fill='lightblue')  # window

    # Car 2
    draw.rectangle([200, 170, 280, 215], fill='blue', outline='darkblue', width=2)
    draw.rectangle([210, 180, 235, 205], fill='lightblue')  # window
    draw.rectangle([245, 180, 270, 205], fill='lightblue')  # window

    # Car 3
    draw.rectangle([320, 185, 380, 220], fill='green', outline='darkgreen', width=2)
    draw.rectangle([330, 195, 350, 215], fill='lightblue')  # window
    draw.rectangle([360, 195, 375, 215], fill='lightblue')  # window

    # Draw trees (another class)
    draw.ellipse([30, 80, 70, 140], fill='darkgreen')
    draw.rectangle([45, 140, 55, 200], fill='brown')

    draw.ellipse([350, 60, 390, 120], fill='darkgreen')
    draw.rectangle([365, 120, 375, 200], fill='brown')

    return np.array(img)

# Create and display sample image
sample_image = create_sample_image()

plt.figure(figsize=(10, 6))
plt.imshow(sample_image)
plt.title('Sample Scene: Road with Multiple Cars and Trees', fontsize=14, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()

print(f"Image shape: {sample_image.shape}")
print(f"Image dtype: {sample_image.dtype}")
```

## Code Example 5.1 (ii) Semantic Segmentation Mask Creation and Visualization

Short Description: This code implements a fundamental semantic segmentation workflow by creating a pixel-wise class mask for a synthetic image. It demonstrates how to label every pixel with a class ID (sky, road, car, tree), visualizes the mask, overlays it on the original image, and analyzes the pixel distribution across classes.

Key Libraries Used:

    numpy: Creates and manipulates the mask array, performs pixel counting and statistical analysis.

    matplotlib.pyplot: Visualizes the original image, the mask, and the overlay in a comparative subplot layout. Uses cmap='tab10' to apply a discrete color map to the class labels.

    PIL (Pillow): The source sample_image (a NumPy array) originated from a PIL Image, demonstrating interoperability.

Code Logic and Flow: High-Level Overview

This script tackles the core task of semantic segmentation: assigning a class label to every pixel in an image. It starts by programmatically generating a ground truth mask based on the known coordinates and shapes of objects in the synthetic scene. The logic defines specific regions of the image array (using NumPy slicing) and fills them with integer codes representing different object classes. It then creates a triple visualization to compare the original image, the pure label mask, and a blended overlay of both. Finally, it performs a quantitative analysis by counting pixels per class, providing insight into the composition of the scene.

Visual Flowchart



Step-by-Step Code Breakdown
Step 1: Programmatic Mask Creation

    Array Initialization: The create_semantic_segmentation_mask function starts by creating a blank 2D NumPy array (mask) of type np.uint8 (integers 0-255), with the same height and width as the input image. This array will hold the class ID for each pixel.

    Class Label Definition: It establishes a mapping of integer IDs to semantic classes: 0 for Sky, 1 for Road, 2 for Car, and 3 for Tree. This is a critical step for any segmentation task.

    Region-Based Assignment: Using NumPy's efficient array slicing, it paints the mask by assigning class IDs to large rectangular regions:

        The top half (mask[0:200, :] = 0) is labeled as Sky.

        The bottom portion (mask[200:, :] = 1) is labeled as Road.

        Specific bounding box coordinates for the three cars are all labeled with class 2. This is a key characteristic of semantic segmentation: it does not distinguish between different instances of the same class.

        Regions for the two trees (including both canopy ellipses and trunk rectangles) are labeled with class 3.

Step 2: Comparative Visualization

    Three-Panel Display: A 1x3 matplotlib subplot is created to display the results side-by-side for effective comparison.

        Panel 1 (Original): Shows the source sample_image.

        Panel 2 (Mask): Displays the semantic_mask array. The cmap='tab10' argument applies a colormap that assigns a distinct, easily readable color to each integer class label (0, 1, 2, 3).

        Panel 3 (Overlay): Creates a blended image to see how well the mask aligns with the original. The overlay is generated by:

            Normalizing the original image to [0, 1].

            Using plt.cm.tab10() to convert the mask integers into RGBA colors, then keeping only RGB.

            Blending 60% of the original image with 40% of the colored mask.

Step 3: Quantitative Analysis

    Pixel Statistics: The code identifies all unique class IDs present in the mask using np.unique().

    Distribution Calculation: For each class, it:

        Creates a boolean array where the mask equals the class ID.

        Counts the True values (pixels) using np.sum().

        Calculates the percentage of total image pixels that belong to that class.

    Summary Output: Prints a clean summary showing the pixel count and percentage for the Sky, Road, Car, and Tree classes. This analysis is similar to evaluating the "coverage" of each object class within the scene.

```python
# 4. Semantic Segmentation Implementation

def create_semantic_segmentation_mask(image_shape):
    """Create semantic segmentation mask - all pixels classified by class only"""
    height, width = image_shape[:2]
    mask = np.zeros((height, width), dtype=np.uint8)

    # Class labels: 0=sky, 1=road, 2=car, 3=tree
    # Sky (background)
    mask[0:200, :] = 0

    # Road
    mask[200:, :] = 1

    # All cars get same label (no instance distinction)
    # Car 1
    mask[180:220, 50:120] = 2
    # Car 2
    mask[170:215, 200:280] = 2
    # Car 3
    mask[185:220, 320:380] = 2

    # Trees
    mask[80:140, 30:70] = 3
    mask[140:200, 45:55] = 3
    mask[60:120, 350:390] = 3
    mask[120:200, 365:375] = 3

    return mask

semantic_mask = create_semantic_segmentation_mask(sample_image.shape)

# Visualize semantic segmentation
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(sample_image)
axes[0].set_title('Original Image', fontsize=12, fontweight='bold')
axes[0].axis('off')

axes[1].imshow(semantic_mask, cmap='tab10')
axes[1].set_title('Semantic Segmentation Mask', fontsize=12, fontweight='bold')
axes[1].axis('off')

# Overlay
overlay = sample_image.copy()
colored_mask = plt.cm.tab10(semantic_mask / 10.0)[:, :, :3]
overlay = (0.6 * overlay/255.0 + 0.4 * colored_mask * 255).astype(np.uint8)
axes[2].imshow(overlay)
axes[2].set_title('Overlay: Image + Semantic Mask', fontsize=12, fontweight='bold')
axes[2].axis('off')

plt.tight_layout()
plt.show()

class_names = {0: 'Sky', 1: 'Road', 2: 'Car', 3: 'Tree'}
unique_classes = np.unique(semantic_mask)
print("Semantic Segmentation Results:")
print(f"Unique classes found: {unique_classes}")
for cls in unique_classes:
    pixel_count = np.sum(semantic_mask == cls)
    percentage = (pixel_count / semantic_mask.size) * 100
    print(f"  Class {cls} ({class_names[cls]}): {pixel_count} pixels ({percentage:.2f}%)")
```

## Code Example 5.1 (iii) Instance Segmentation Mask Creation and Multi-Object Analysis

Short Description: This code implements instance segmentation by creating unique masks for each individual object in a synthetic scene. It distinguishes between different instances of the same class (three separate cars, two separate trees), generates combined and individual visualizations, and provides detailed analytics including pixel counts and bounding box coordinates for each instance.

Key Libraries Used:

    numpy: Creates and manipulates binary mask arrays for each object instance and calculates pixel statistics

    matplotlib.pyplot: Generates a comprehensive 2x3 grid visualization showing multiple perspectives of instance segmentation

    PIL.ImageDraw: Draws colored bounding boxes and labels directly onto the original image for instance identification

Code Logic and Flow: High-Level Overview

This script advances beyond semantic segmentation by implementing instance segmentation, where each individual object receives a unique identifier regardless of its class. The core logic involves creating separate binary masks for each object instance (3 cars, 2 trees), storing them in a structured format with metadata, and then combining them into a unified visualization. The approach demonstrates how instance segmentation provides more granular information than semantic segmentation, enabling applications that require tracking, counting, or manipulating individual objects within the same class.

Visual Flowchart



Step-by-Step Code Breakdown
Step 1: Instance-Specific Mask Creation

    Individual Binary Masks: The create_instance_segmentation_masks function creates five separate binary (0/1) mask arrays, one for each object instance in the synthetic scene. Unlike semantic segmentation where all cars share one label, here each car and tree gets its own unique identifier.

    Structured Data Storage: Each instance is stored as a dictionary containing:

        class: The semantic category ('car' or 'tree')

        instance_id: A unique integer identifier (1-5)

        mask: The binary array marking the object's pixels

        bbox: The bounding box coordinates in [x_min, y_min, x_max, y_max] format

    Multi-Part Object Handling: For trees (which consist of canopy + trunk), the function correctly combines both regions into a single binary mask, demonstrating how instance segmentation handles objects composed of multiple disconnected components.

Step 2: Comprehensive Visualization Grid

    2×3 Multi-Panel Display: Creates a sophisticated visualization with six subplots arranged in two rows and three columns to convey different aspects of instance segmentation:

        Row 1, Plot 1: Original synthetic image for reference

        Row 1, Plot 2: Combined instance mask where each instance has a unique color (using cmap='tab10'). This directly contrasts with semantic segmentation by showing different colors for each car

        Row 1, Plot 3: Original image overlaid with colored bounding boxes and text labels (e.g., "car 1", "tree 2") using PIL.ImageDraw

        Row 2, Plots 1-3: Individual binary masks for the first three instances (the three cars), showing the raw pixel-level segmentation for each object

Step 3: Instance Combination and Analysis

    Unified Instance Map: Creates a combined_instance_mask array by iterating through all instances and assigning each object's pixels its unique instance_id. This produces a single array where pixel values represent instance membership rather than class membership.

    Quantitative Analytics: For each instance, calculates:

        Pixel Count: Using np.sum(inst['mask']) to count the number of pixels belonging to that specific object

        Bounding Box: The stored [x_min, y_min, x_max, y_max] coordinates that tightly enclose the object

    Summary Output: Prints a structured report showing all five instances with their class, ID, pixel count, and bounding box coordinates. This demonstrates the richer information available from instance segmentation compared to semantic segmentation.

Step 4: Instance Segmentation vs. Semantic Segmentation

    Granular Distinction: Highlights the key difference between the two segmentation types: instance segmentation provides object-level identification (car 1, car 2, car 3) while semantic segmentation provides only class-level identification (car, car, car).

    Practical Applications: The visualization and analytics demonstrate why instance segmentation is essential for tasks requiring individual object manipulation, counting, or tracking, such as autonomous vehicle perception, inventory management, or biological cell analysis.

```python
# 5. Instance Segmentation Implementation

def create_instance_segmentation_masks(image_shape):
    """Create instance segmentation masks - each object instance has unique ID"""
    height, width = image_shape[:2]

    # Create separate masks for each instance
    instances = []

    # Instance 1: Car 1 (red car)
    mask1 = np.zeros((height, width), dtype=np.uint8)
    mask1[180:220, 50:120] = 1
    instances.append({'class': 'car', 'instance_id': 1, 'mask': mask1, 'bbox': [50, 180, 120, 220]})

    # Instance 2: Car 2 (blue car)
    mask2 = np.zeros((height, width), dtype=np.uint8)
    mask2[170:215, 200:280] = 1
    instances.append({'class': 'car', 'instance_id': 2, 'mask': mask2, 'bbox': [200, 170, 280, 215]})

    # Instance 3: Car 3 (green car)
    mask3 = np.zeros((height, width), dtype=np.uint8)
    mask3[185:220, 320:380] = 1
    instances.append({'class': 'car', 'instance_id': 3, 'mask': mask3, 'bbox': [320, 185, 380, 220]})

    # Instance 4: Tree 1
    mask4 = np.zeros((height, width), dtype=np.uint8)
    mask4[80:140, 30:70] = 1
    mask4[140:200, 45:55] = 1
    instances.append({'class': 'tree', 'instance_id': 4, 'mask': mask4, 'bbox': [30, 80, 70, 200]})

    # Instance 5: Tree 2
    mask5 = np.zeros((height, width), dtype=np.uint8)
    mask5[60:120, 350:390] = 1
    mask5[120:200, 365:375] = 1
    instances.append({'class': 'tree', 'instance_id': 5, 'mask': mask5, 'bbox': [350, 60, 390, 200]})

    return instances

instances = create_instance_segmentation_masks(sample_image.shape)

# Visualize instance segmentation
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes[0, 0].imshow(sample_image)
axes[0, 0].set_title('Original Image', fontsize=12, fontweight='bold')
axes[0, 0].axis('off')

# Combined instance mask
combined_instance_mask = np.zeros(sample_image.shape[:2], dtype=np.uint8)
for inst in instances:
    combined_instance_mask[inst['mask'] > 0] = inst['instance_id']

axes[0, 1].imshow(combined_instance_mask, cmap='tab10')
axes[0, 1].set_title('Instance Segmentation Mask\n(Each color = unique instance)', fontsize=12, fontweight='bold')
axes[0, 1].axis('off')

# Overlay with bounding boxes
overlay_img = Image.fromarray(sample_image)
draw = ImageDraw.Draw(overlay_img)
colors = ['red', 'blue', 'green', 'orange', 'purple']
for i, inst in enumerate(instances):
    bbox = inst['bbox']
    draw.rectangle(bbox, outline=colors[i], width=3)
    draw.text((bbox[0], bbox[1]-15), f"{inst['class']} {inst['instance_id']}", fill=colors[i])

axes[0, 2].imshow(overlay_img)
axes[0, 2].set_title('Instance Segmentation with Bounding Boxes', fontsize=12, fontweight='bold')
axes[0, 2].axis('off')

# Show individual instance masks
for i in range(3):
    if i < len(instances):
        axes[1, i].imshow(instances[i]['mask'], cmap='gray')
        axes[1, i].set_title(f"Instance {instances[i]['instance_id']}: {instances[i]['class']}", fontsize=10, fontweight='bold')
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()

print("Instance Segmentation Results:")
print(f"Total instances detected: {len(instances)}")
for inst in instances:
    pixel_count = np.sum(inst['mask'])
    print(f"  Instance {inst['instance_id']} ({inst['class']}): {pixel_count} pixels, BBox: {inst['bbox']}")
```

## Code Example 5.1 (v) Mask R-CNN Workflow Simulation: A Step-by-Step Architectural Breakdown

It demonstrates the sequential process from feature extraction to final instance mask prediction, highlighting how the model combines object detection with pixel-wise segmentation through specialized components like Region Proposal Networks (RPN) and ROI Align.

    scipy.ndimage.zoom: Simulates the downsampling of feature maps and the resizing operations for Region of Interest (RoI) alignment.

Code Logic and Flow: High-Level Overview

This exercise provides a conceptual walkthrough of the Mask R-CNN architecture by implementing a simulated, non-trainable version of its pipeline. The core logic follows the model's classic, sequential stages: a backbone CNN (like ResNet) extracts hierarchical features; a Region Proposal Network (RPN) suggests potential object regions; RoI Align pools features from these regions into a fixed size; and finally, parallel prediction heads output the class label, refined bounding box, and a binary mask for each instance. By simulating each step with pre-defined outputs for the known synthetic image, the code demystifies the data flow and intermediate representations within this influential architecture.

Visual Flowchart



Step-by-Step Code Breakdown
Step 1: Backbone Feature Extraction

    Architectural Role: The backbone_feature_extraction method simulates the first stage of the pipeline, typically performed by a deep CNN like ResNet, possibly with a Feature Pyramid Network (FPN). The backbone's job is to convert the raw pixel image into a set of rich, hierarchical feature maps that encode patterns useful for detection.

    Simulation Technique: Since this is a simulation, it uses scipy.ndimage.zoom to downsample the image by a factor of 4 (zoom factor = 0.25), reducing its spatial dimensions while preserving 3 color channels. This crude approximation represents the semantic feature maps a real backbone would produce.

    Purpose: This step highlights the transition from pixel space to feature space, where subsequent components (RPN, prediction heads) operate.

Step 2: Region Proposal Network (RPN)

    Core Function: The region_proposal_network method simulates the RPN, a lightweight network that scans the feature maps to propose regions (proposals) that likely contain objects. In a real RPN, this involves sliding a small network over the feature map to output "objectness" scores and bounding box adjustments for a set of anchor boxes at each location.

    Simulated Output: The method returns a hardcoded list of proposals matching the known objects in our synthetic image. Each proposal includes coordinates (bbox) and an objectness score, representing the RPN's confidence that the region contains any foreground object (not yet a specific class).

Step 3: RoI Align and Feature Pooling

    Critical Operation: The roi_align function is a simplified version of the actual RoI Align operation. Its purpose is to extract a small, fixed-size feature map (e.g., 14x14) from the backbone's feature maps for each region proposal, regardless of the proposal's original size or aspect ratio.

    Why it Matters: Earlier methods like RoI Pooling used coarse quantization, harming mask accuracy. RoI Align uses bilinear interpolation to precisely align the extracted features, which is crucial for predicting accurate pixel masks. The simulation achieves this resizing using scipy.ndimage.zoom.

Step 4: Parallel Prediction Heads

    Three-Task Design: The prediction_heads method simulates the three parallel heads that take the aligned RoI features (roi_features) as input:

        A classification head predicts the object's class (e.g., 'car', 'tree').

        A bounding box regression head refines the proposal's coordinates for a tighter fit.

        A mask head, typically a small Fully Convolutional Network (FCN), predicts a binary segmentation mask within the bounding box.

    Simulation Integration: For this demo, the class and score are taken from the RPN step, and the mask is fetched from the instances list created in the previous instance segmentation exercise. In a real Mask R-CNN, all three outputs are predicted simultaneously from the shared RoI features.

Step 5: Comprehensive Workflow Visualization

    Multi-Panel Narrative: The visualize_workflow method creates a detailed figure using gridspec to tell the story of Mask R-CNN:

        Input and Features: Shows the original image and the downsampled "feature maps."

        RPN Output: Displays the image with yellow proposal boxes drawn on it.

        Final Detections: For each detection, it shows the original image with the instance's mask overlaid in red and the final predicted bounding box in green, along with the class label and score.

    Educational Value: This visualization powerfully illustrates the transformation from pixels to proposals and finally to precise instance segmentations, connecting the conceptual stages to tangible outputs.

```python
# 7. Mask R-CNN Workflow Simulation

class SimplifiedMaskRCNN:
    """Simplified Mask R-CNN workflow demonstration"""

    def __init__(self, image):
        self.image = image
        self.feature_maps = None
        self.proposals = None
        self.detections = None

    def backbone_feature_extraction(self):
        """Step 1: Backbone CNN extracts feature maps (simulated)"""
        print("Step 1: Backbone Feature Extraction (e.g., ResNet)")
        # Simulate feature extraction with downsampling
        from scipy.ndimage import zoom
        self.feature_maps = zoom(self.image, (0.25, 0.25, 1), order=1)
        print(f"  Input shape: {self.image.shape}")
        print(f"  Feature map shape: {self.feature_maps.shape}")
        return self.feature_maps

    def region_proposal_network(self):
        """Step 2: RPN proposes object bounding boxes"""
        print("\nStep 2: Region Proposal Network (RPN)")
        # Simulate RPN proposals
        self.proposals = [
            {'bbox': [50, 180, 120, 220], 'score': 0.95},
            {'bbox': [200, 170, 280, 215], 'score': 0.92},
            {'bbox': [320, 185, 380, 220], 'score': 0.89},
            {'bbox': [30, 80, 70, 200], 'score': 0.87},
            {'bbox': [350, 60, 390, 200], 'score': 0.85}
        ]
        print(f"  Proposals generated: {len(self.proposals)}")
        for i, prop in enumerate(self.proposals):
            print(f"    Proposal {i+1}: BBox {prop['bbox']}, Score: {prop['score']:.2f}")
        return self.proposals

    def roi_align(self, bbox):
        """Step 3: RoI Align extracts fixed-size features per box"""
        x1, y1, x2, y2 = bbox
        roi_features = self.image[y1:y2, x1:x2]
        # Resize to fixed size (e.g., 14x14)
        from scipy.ndimage import zoom
        target_size = (14, 14)
        zoom_factors = (target_size[0]/roi_features.shape[0],
                       target_size[1]/roi_features.shape[1], 1)
        roi_aligned = zoom(roi_features, zoom_factors, order=1)
        return roi_aligned

    def prediction_heads(self):
        """Step 4: Parallel heads predict class, bbox, and mask"""
        print("\nStep 3 & 4: RoI Align + Prediction Heads")
        self.detections = []
        class_names = ['car', 'car', 'car', 'tree', 'tree']

        for i, prop in enumerate(self.proposals):
            # RoI Align
            roi_features = self.roi_align(prop['bbox'])

            # Simulate predictions
            detection = {
                'bbox': prop['bbox'],
                'class': class_names[i],
                'class_score': prop['score'],
                'mask': instances[i]['mask'],
                'roi_features_shape': roi_features.shape
            }
            self.detections.append(detection)

            print(f"  Detection {i+1}:")
            print(f"    RoI features shape: {roi_features.shape}")
            print(f"    Predicted class: {detection['class']} (score: {detection['class_score']:.2f})")
            print(f"    Mask shape: {detection['mask'].shape}")

        return self.detections

    def visualize_workflow(self):
        """Visualize the complete Mask R-CNN workflow"""
        fig = plt.figure(figsize=(16, 10))
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

        # Original image
        ax1 = fig.add_subplot(gs[0, 0])
        ax1.imshow(self.image)
        ax1.set_title('Input Image', fontweight='bold')
        ax1.axis('off')

        # Feature maps
        ax2 = fig.add_subplot(gs[0, 1])
        ax2.imshow(self.feature_maps)
        ax2.set_title(f'Feature Maps\n{self.feature_maps.shape}', fontweight='bold')
        ax2.axis('off')

        # RPN proposals
        ax3 = fig.add_subplot(gs[0, 2])
        img_with_proposals = Image.fromarray(self.image)
        draw = ImageDraw.Draw(img_with_proposals)
        for prop in self.proposals:
            draw.rectangle(prop['bbox'], outline='yellow', width=2)
        ax3.imshow(img_with_proposals)
        ax3.set_title(f'RPN Proposals\n({len(self.proposals)} boxes)', fontweight='bold')
        ax3.axis('off')

        # Individual detections with masks
        for i in range(min(6, len(self.detections))):
            row = 1 + i // 3
            col = i % 3
            ax = fig.add_subplot(gs[row, col])

            # Show mask overlay
            mask_overlay = self.image.copy()
            mask_colored = np.zeros_like(self.image)
            mask_colored[self.detections[i]['mask'] > 0] = [255, 0, 0]
            mask_overlay = (0.7 * mask_overlay + 0.3 * mask_colored).astype(np.uint8)

            ax.imshow(mask_overlay)
            bbox = self.detections[i]['bbox']
            rect = plt.Rectangle((bbox[0], bbox[1]), bbox[2]-bbox[0], bbox[3]-bbox[1],
                                linewidth=2, edgecolor='lime', facecolor='none')
            ax.add_patch(rect)
            ax.set_title(f"Instance {i+1}: {self.detections[i]['class']}\nScore: {self.detections[i]['class_score']:.2f}",
                        fontsize=10, fontweight='bold')
            ax.axis('off')

        plt.suptitle('Mask R-CNN Workflow: From Image to Instance Masks',
                    fontsize=16, fontweight='bold', y=0.98)
        plt.show()

# Run Mask R-CNN simulation
print("="*60)
print("MASK R-CNN WORKFLOW SIMULATION")
print("="*60)

mask_rcnn = SimplifiedMaskRCNN(sample_image)
mask_rcnn.backbone_feature_extraction()
mask_rcnn.region_proposal_network()
mask_rcnn.prediction_heads()
mask_rcnn.visualize_workflow()

print("\n" + "="*60)
print("Mask R-CNN workflow completed successfully!")
print("="*60)
```

```python
# 8. DeepSeek API Integration for Segmentation Analysis

def analyze_segmentation_with_deepseek(semantic_results, instance_results):
    """Use DeepSeek API to analyze and compare segmentation results"""

    prompt = f"""Analyze the following computer vision segmentation results:

SEMANTIC SEGMENTATION RESULTS:
- Total unique classes: {len(np.unique(semantic_mask))}
- Class distribution:
  * Sky: {np.sum(semantic_mask == 0)} pixels
  * Road: {np.sum(semantic_mask == 1)} pixels
  * Car (all instances): {np.sum(semantic_mask == 2)} pixels
  * Tree (all instances): {np.sum(semantic_mask == 3)} pixels

INSTANCE SEGMENTATION RESULTS:
- Total instances detected: {len(instances)}
- Car instances: {sum(1 for inst in instances if inst['class'] == 'car')}
- Tree instances: {sum(1 for inst in instances if inst['class'] == 'tree')}

Provide a brief analysis (3-4 sentences) explaining:
1. The key difference between these two segmentation approaches
2. Which approach would be better for counting objects
3. A real-world application where instance segmentation is essential"""

    payload = {
        "model": "deepseek-chat",
        "messages": [
            {"role": "system", "content": "You are a computer vision expert specializing in image segmentation techniques."},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.7,
        "max_tokens": 300
    }

    try:
        response = requests.post(DEEPSEEK_API_URL, headers=headers, json=payload, timeout=30)
        response.raise_for_status()
        result = response.json()
        analysis = result['choices'][0]['message']['content']
        return analysis
    except Exception as e:
        return f"Error calling DeepSeek API: {str(e)}"

print("="*60)
print("DEEPSEEK API ANALYSIS")
print("="*60)
print("\nQuerying DeepSeek API for segmentation analysis...\n")

analysis = analyze_segmentation_with_deepseek(semantic_mask, instances)
print("DeepSeek Analysis:")
print("-" * 60)
print(analysis)
print("-" * 60)
```

```python
# 9. Performance Metrics and Comparison Table

def calculate_segmentation_metrics(semantic_mask, instances):
    """Calculate and compare metrics for both segmentation types"""

    metrics = {
        'semantic': {
            'output_shape': semantic_mask.shape,
            'unique_labels': len(np.unique(semantic_mask)),
            'memory_size_kb': semantic_mask.nbytes / 1024,
            'can_count_instances': False,
            'complexity': 'Low',
            'inference_speed': 'Fast'
        },
        'instance': {
            'output_shape': f"{len(instances)} masks of {instances[0]['mask'].shape}",
            'unique_instances': len(instances),
            'memory_size_kb': sum(inst['mask'].nbytes for inst in instances) / 1024,
            'can_count_instances': True,
            'complexity': 'High',
            'inference_speed': 'Slower'
        }
    }

    return metrics

metrics = calculate_segmentation_metrics(semantic_mask, instances)

print("\n" + "="*60)
print("PERFORMANCE METRICS COMPARISON")
print("="*60)

comparison_data = [
    ['Metric', 'Semantic Segmentation', 'Instance Segmentation'],
    ['-'*20, '-'*25, '-'*25],
    ['Output Format', f"{metrics['semantic']['output_shape']}", f"{metrics['instance']['output_shape']}"],
    ['Unique Labels/Instances', f"{metrics['semantic']['unique_labels']}", f"{metrics['instance']['unique_instances']}"],
    ['Memory Usage', f"{metrics['semantic']['memory_size_kb']:.2f} KB", f"{metrics['instance']['memory_size_kb']:.2f} KB"],
    ['Can Count Objects', f"{metrics['semantic']['can_count_instances']}", f"{metrics['instance']['can_count_instances']}"],
    ['Model Complexity', f"{metrics['semantic']['complexity']}", f"{metrics['instance']['complexity']}"],
    ['Inference Speed', f"{metrics['semantic']['inference_speed']}", f"{metrics['instance']['inference_speed']}"],
]

for row in comparison_data:
    print(f"{row[0]:<25} {row[1]:<25} {row[2]:<25}")

# Visualization of metrics
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Memory comparison
memory_data = [metrics['semantic']['memory_size_kb'], metrics['instance']['memory_size_kb']]
axes[0].bar(['Semantic', 'Instance'], memory_data, color=['skyblue', 'coral'])
axes[0].set_ylabel('Memory Usage (KB)', fontweight='bold')
axes[0].set_title('Memory Footprint Comparison', fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)

# Object counting capability
counting_data = [0, len(instances)]
axes[1].bar(['Semantic\n(Cannot distinguish)', 'Instance\n(Can count each)'],
            counting_data, color=['lightgray', 'lightgreen'])
axes[1].set_ylabel('Number of Distinguishable Objects', fontweight='bold')
axes[1].set_title('Object Counting Capability', fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("Analysis complete!")
print("="*60)
```
